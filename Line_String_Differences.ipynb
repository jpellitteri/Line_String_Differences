```python
#Final Project for Data Representations Course | Coursera

def singleline_diff(line1, line2):
    """
    Inputs:
        line1 - first single line string
        line2 - second single line string
    Outputs:
        Returns the index where the first difference between
        line1 and line2 occurs.
        
        Returns IDENTICAL if the two lines are the same.
    
    """
    # variables assigned for length of each line and IDENTICAL
    len_1 = len(line1) 
    len_2 = len(line2) 
    IDENTICAL = -1
    # return IDENTICAL if both lines are identical
    if line1 == line2: 
        return IDENTICAL
    
    # loop function to compare equal index between line within range of longest line
    for i in range(max(len_1, len_2)):
    # if values are not equal and less than smallest length, return that index i
        if i < min(len_1, len_2):
            if line1[i] != line2[i]:
                return i
    # if i is greater than the shortest length of lines then return value of index i
        elif i >= min(len_1, len_2):
            return min(len_1, len_2)
        

            
```

```python
# function for formatting output of line differences

def singleline_diff_format(line1, line2, idx):
    """
    Inputs:
      line1 - first single line string
      line2 - second single line string
      idx   - index at which to indicate difference
    Output:
      Returns a three line formatted string showing the location
      of the first difference between line1 and line2.

      If either input line contains a newline or carriage return,
      then returns an empty string.

      If idx is not a valid index, then returns an empty string.
    """
   
    # if there is a newline in any line or if idx is negative, then return empty string
    if ("\n" or "\r") in line1 or (("\n" or "\r") in line2) or \
    int(idx) < 0:
        format = ""
        return format
    # if idx is outside range of shortest length of lines return empty string
    elif (len(line1) < len(line2)) and (int(idx) > len(line1)):
        format = ""
        return format
    elif (len(line2) <= len(line1)) and (int(idx) > len(line2)):
        format = ""
        return format
    
    # if idx is the first index then return "^" to reference the first index
    else:
        if int(idx) == 0:
            s = "^"
            format = line1 + '\n' + s + '\n' + line2 + '\n'
            return format
    # if idx is greater than 0 then return "=" times the idx and then "^"
        else:
            s = ("=")*(int(idx)) + "^"
            format = line1 + '\n' + s + '\n' + line2 + '\n'
            return format

            

            
            
            
        
    
                   
    
                
                
                
            
    
    
    
    
```

```python

IDENTICAL = -1

def multiline_diff(lines1, lines2):
    """
    Inputs:
      lines1 - list of single line strings
      lines2 - list of single line strings
    Output:
      Returns a tuple containing the line number (starting from 0) and
      the index in that line where the first difference between lines1
      and lines2 occurs.
      
      Returns (IDENTICAL, IDENTICAL) if the two lists are the same.
    """
    
    # if all lines are identical, return IDENTICAL
    if lines1 == lines2:
        return (IDENTICAL, IDENTICAL)
    
    # loop function for the lines in each list (lines1 and lines2)
    for index in range(min(len(lines1), len(lines2))):
    # assign each line as an index of the lists
        line1 = lines1[index]
        line2 = lines2[index]
    # use previous function to assign first different character
        idx = singleline_diff(line1, line2)
        if idx >= 0:
            if lines1[index] != lines2[index]:
                return (index, idx)
    # if the lists are different length, return the minimum length and the first character
    if len(lines1) != len(lines2):
        return (min(len(lines1), len(lines2)), 0)
                
```

```python


# output lines of a file without newline characters

def get_file_lines(filename):
    """
    Inputs:
      filename - name of file to read
    Output:
      Returns a list of lines from the file named filename.  Each
      line will be a single line string with no newline ('\n') or 
      return ('\r') characters.

      If the file does not exist or is not readable, then the
      behavior of this function is undefined.
    """
    
   # opens and reads the list of lines without newlines "\n" or "\r" 
    with open(filename, 'rt') as open_file:
        file_lines = open_file.read().splitlines()
        return file_lines
    
```

```python
# output difference between two files with formatting

def file_diff_format(filename1, filename2):
    """
    Inputs:
      filename1 - name of first file
      filename2 - name of second file
    Output:
      Returns a four line string showing the location of the first
      difference between the two files named by the inputs.

      If the files are identical, the function instead returns the
      string "No differences\n".

      If either file does not exist or is not readable, then the
      behavior of this function is undefined.
    """
    
    # if files are identical, return "No differences\n"
    if filename1 == filename2:
        return "No differences\n"
    
    # use previous function to output list of lines from each file and assigns them
    lines1 = get_file_lines(filename1)
    lines2 = get_file_lines(filename2)
    
    # apply previous multi_line function on list of strings
    line_diff = multiline_diff(lines1, lines2)
    # assigns the line and character where the difference is
    line_num = line_diff[0]
    str_idx = line_diff[1]
    line_idx = singleline_diff_format(lines1[line_num], lines2[line_num], str_idx)
    
        
    return "Line" + " " + str(line_num)+":" + "\n" + line_idx
    
    
    
    
    
```
